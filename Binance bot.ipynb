```python
import asyncio
from binance import AsyncClient, BinanceSocketManager

async def main():
    client = await AsyncClient.create()
    bm = BinanceSocketManager(client)
    # start any sockets here, i.e a trade socket
    ts = bm.trade_socket('BNBBTC')
    # then start receiving messages
    async with ts as tscm:
        while True:
            res = await tscm.recv()
            print(res)

    await client.close_connection()



# Define the cryptocurrency symbol
# symbol = 'BTCUSDT'  # Bitcoin against USDT (Tether)

# # Initialize empty lists to store prices and timestamps
# prices = []
# timestamps = []

# # Function to process incoming messages
# def process_message(msg):
#     timestamp = msg['k']['t']
#     formatted_time = datetime.fromtimestamp(timestamp / 1000).strftime('%Y-%m-%d %H:%M:%S')
#     price = float(msg['k']['c'])

#     prices.append(price)
#     timestamps.append(formatted_time)

#     plt.plot(timestamps, prices)
#     plt.title('Live Crypto Price Updates')
#     plt.xlabel('Time')
#     plt.ylabel('Price (USDT)')
#     plt.xticks(rotation=45, ha='right')
#     plt.pause(1e-10)  # Update the plot

# # Start the WebSocket for the symbol
# conn_key = bm.start_kline_socket(symbol, process_message)

# # Display the plot
# plt.ion()
# plt.show()

# # Start the WebSocket manager
# bm.start()
```
