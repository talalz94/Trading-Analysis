```python
import pandas as pd

def view_tag(df_feat: pd.DataFrame, specs, tag: str, n: int = 30):
    spec = next(s for s in specs if s.tag == tag)
    cols = spec.cols

    base = ["t", "open", "high", "low", "close", "volume"]
    keep = [c for c in base if c in df_feat.columns] + cols

    out = df_feat[keep].copy()

    # nice ordering: base first, then indicator cols
    out = out.loc[:, keep]

    return out.tail(n), cols

tag_df, tag_cols = view_tag(df_feat, specs, tag="rsi14", n=50)
#tag_cols, tag_df
```

```python
def inspect_tag(df_feat, specs, tag: str):
    spec = next(s for s in specs if s.tag == tag)
    cols = spec.cols

    info = pd.DataFrame({
        "col": cols,
        "dtype": [df_feat[c].dtype for c in cols],
        "missing_%": [(df_feat[c].isna().mean() * 100) for c in cols],
        "last_value": [df_feat[c].dropna().iloc[-1] if df_feat[c].notna().any() else None for c in cols],
    }).sort_values("missing_%", ascending=True)

    return info, df_feat[cols].describe(include="all")

info, desc = inspect_tag(df_feat, specs, "rsi14")
info, desc
```
